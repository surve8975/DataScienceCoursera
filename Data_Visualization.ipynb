```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

plt.plot(x,y, 'r*')
plt.legend()
plt.text(2,5, 'Comments can be written here')
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [5, 3, 1, 4, 2]

plt.plot(x1, y1, label = 'Data 1')
plt.plot(x2, y2, label = 'Data 2')
plt.title("Main title")
plt.xlabel('x label')
plt.ylabel('y label')
plt.xlim(-2, 8)
plt.legend(title = 'legend title')
```

```python
df = sns.load_dataset('iris')
```

```python
df.groupby('species').count()
```

```python
df.groupby('species').sum()
```

```python
df = df[df['species']== 'setosa']
plt.scatter(df['sepal_length'], df['sepal_width'])
```

```python
df.groupby('species').sum()
```

```python
for species, group in df.groupby('species'):
      plt.scatter(group['sepal_length'],group['sepal_width'] , label =species)

plt.legend()
plt.show()
```

```python
plt.scatter(df['sepal_length'], df['sepal_width'])
```

```python
plt.hist(df['sepal_length'], bins = 15, color = 'skyblue', edgecolor = 'black')
plt.show()
```

```python
sns.boxplot(x = 'sepal_width', y = 'species', data = df)
```
